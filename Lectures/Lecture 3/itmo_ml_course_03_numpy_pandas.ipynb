#### ndarray

Массив — упорядоченная изменяемая структура данных, к элементам которой можно обратиться по индексу

Основным объектом NumPy является **ndarray** (n-dimensional array) - структура данных, которая позволяет хранить набор элементов одного типа

Массивы NumPy имеют фиксированный размер при создании, в отличие от списков Python (которые могут динамически увеличиваться). Изменение размера ndarray создаст новый массив и удалит исходный.

#### Соединение/разбиение массивов

```python
A = np.array([[1, 1],
              [0, 1]])
B = np.array([[2, 0],
              [3, 4]])

vstack_result = np.vstack((A, B))
vstack_result
```

```python
hstack_result = np.hstack((A, B))
hstack_result
```

```python
np.vsplit(vstack_result, 2)
```

```python
np.hsplit(hstack_result, 2)
```

Скорость вычислений и strides

```python
x = np.zeros((1000,))
y = np.zeros((200000,)).reshape((1000, -1))[:, 0]
x.shape, y.shape
```

```python
%%timeit

x.sum()
```

```python
%%timeit

y.sum()
```

```python
x.strides, y.strides
```

#### Основные структуры

**Series**

Одномерный массив с метками, способный хранить данные любого типа (целые числа, строки, числа с плавающей запятой, объекты Python и т. д.). Метки осей вместе называются индексом.

```python
s = pd.Series(np.arange(5), index=["a", "b", "c", "d", "e"])
s
```

```python
s['d']
```

### Join

```python
left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"]
)
left
```

```python
right = pd.DataFrame(
    {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
)
right
```

```python
left.join(right)
```
