## Plotly

Plotly -  библиотека для интерактивной визуализации данных

[Документация](https://plotly.com/python/)



```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_excel('Направление.xlsx')
df.head()
```

```python
df = df.dropna()
```

#### Sunburst plot - визуализация иерархических данных

```python
fig = px.sunburst(df, path=['Destination', 'Traffic group', 'Traffic subgroup'])
fig.show()
```

#### Icicle - визуализация иерархических данных

```python
fig = px.icicle(df, path=[px.Constant("all"), 'Destination', 'Traffic group', 'Traffic subgroup'])
fig.update_traces(root_color="lightgrey")
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

### Boxplot

```python
df = pd.read_excel('Реестр_выполненных_рейсов_за_период.xlsx', header=1).drop(columns=['Unnamed: 0', 'Unnamed: 13', 'Unnamed: 15'])
df.head()
```

```python
df = df.dropna()
df = df[df['В/П'] !='Spot Number']
```

```python
fig = px.box(df, x="Направление", y="Вместимость", color="В/П")
fig.show()
```

```python
fig = px.scatter(
    df, x='Вместимость', y='Baggage Weight', opacity=0.65,
    trendline='ols', trendline_color_override='darkblue'
)
fig.show()
```

#### Funnel chart - воронка

```python
df = pd.DataFrame(dict(
    stage = ['Sent', 'Received', 'Clicks', 'Sales', 'Sent', 'Received', 'Clicks', 'Sales'],
    percentage = [70, 32, 20, 4, 30, 38, 10, 6],
    campaign = ['C1', 'C1', 'C1', 'C1', 'C2', 'C2', 'C2', 'C2']))
df.head()
```

```python
fig = px.funnel(df, x='percentage', y='stage', color='campaign')
fig.show()
```

#### Heatmap на карте

Набор данных представляет собой данные переписи населения Галисии - автономного сообщества Испании. Координаты - центр каждого участка переписи.

```python
df = pd.read_csv('https://raw.githubusercontent.com/R-CoderDotCom/data/main/sample_datasets/population_galicia.csv')
df.sample(n=10).sort_values('tot_pob', ascending=False)
```

```python
fig = px.density_mapbox(
    df,
    lat='latitude',
    lon='longitude',
    z='tot_pob',
    radius=7,
    center=dict(lat=42.83, lon=-8.35),
    zoom=6,
    mapbox_style='open-street-map',
    color_continuous_scale='rainbow',
    opacity=0.5,
)
fig.show()
```

```python
fig.write_html("map.html")
```
